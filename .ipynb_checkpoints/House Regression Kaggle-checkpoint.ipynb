```python
# Impute Missing values for missing data

X = df.drop(columns="SalePrice", axis=1)
y = df["SalePrice"]

X_numeric = X.select_dtypes(exclude="object")
X_numeric
```

```python
from sklearn.impute import SimpleImputer

simple_imputer = SimpleImputer(strategy="mean")
X_numeric_imputed = pd.DataFrame(simple_imputer.fit_transform(X_numeric), columns=X_numeric.columns)
X_numeric_imputed
```

```python
# find all the categorical data
X_categorical = X.select_dtypes(include="object")
X_categorical
```

```python
#See if it has any missing values
X_categorical.isnull().sum()
```

```python
# List of features with more than 100 missing data
X_categorical.shape[0] - X_categorical.isnull().sum() < 50

# Drops the feature with the highest amount of missing data
X_categorical.drop(columns=["Fence"], inplace=True)
```

```python
# fill in unknowns with the highest occurence categorical value 
X_categorical = X_categorical.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
low_cardanility = [col for col in X_categorical.columns if X_categorical[col].nunique() < 6]
high_cardanility = [col for col in X_categorical.columns if X_categorical[col].nunique() >= 6]
print(f" Low cardanility:\n {np.array(low_cardanility)}")
print(f"\nHigh cardanility:\n {np.array(high_cardanility)}")
```

```python
X_one_hot_encoded = pd.get_dummies(X_categorical[low_cardanility])
X_one_hot_encoded
```

```python

from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

#Implement ordinal encoding
X_ordinal_encoded = pd.DataFrame(ordinal_encoder.fit_transform(X_categorical[high_cardanility]), columns=high_cardanility)
X_ordinal_encoded
```

```python
# Combine the pandas dataset together
X_clean = pd.concat([X_ordinal_encoded, X_one_hot_encoded, X_numeric_imputed], axis=1)
X_clean
```
